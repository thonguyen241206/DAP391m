```python
import pandas as pd
titanic = pd.read_csv("titanic.csv")
#1
embarked_map = {
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'}
titanic['embarked'] = titanic['embarked'].map(embarked_map).astype('category')
print(titanic.embarked.unique())
print(titanic.embarked.value_counts())
#2
sex_map = {
    0: 'Male',
    1: 'Female'}
titanic['sex'] = titanic['sex'].map(sex_map).astype('category')
print(titanic.sex.unique())
print(titanic.sex.value_counts())
#3
survived_map = {
    0: 'No',
    1: 'Yes'}
titanic['survived'] = titanic['survived'].map(survived_map).astype('category')
print("Survived conversion complete:")
print(titanic.survived.unique())
print(titanic.survived.value_counts())
print("-" * 30)

#4
pclass_map = {
    1: '1st',
    2: '2nd',
    3: '3rd'}
titanic['pclass'] = titanic['pclass'].map(pclass_map)
pclass_type = pd.CategoricalDtype(categories=['1st', '2nd', '3rd'], ordered=True)
titanic['pclass'] = titanic['pclass'].astype(pclass_type)

display(titanic.head())
print(titanic.pclass.unique())
print(titanic.pclass.value_counts())
# Check the order of pclass as requested
print("So s√°nh: '1st' < '3rd'", titanic['pclass'].iloc[0] < '3rd')
```
